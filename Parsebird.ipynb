```python
def clean_month(month_list):
    full_month_list = []
    for m in month_list:    
        if m == "":
            full_month_list.append(prior_month)
        else:
            full_month_list.append(m)
            prior_month = m
    return full_month_list
```

```python
def get_records_from_file (filename):
    all_tokens = []
    with open(filename, 'r') as f:
        for line in f.readlines():
            tokens = line.strip().split('\t')   # tokenize the row based on spaces
            all_tokens.append(tokens)
            

    db_records = []
    location = filename[7:-4]
    month_list = clean_month(all_tokens[MONTH_INDEX])
    for bird_index in range(BIRD_START_INDEX, len(all_tokens)-1):
        bird_data = all_tokens[bird_index]
        for bird_record_index in range(1, len(bird_data)):
            if bird_record_index < len(month_list):
                db_records.append ((location, bird_data[0], month_list[bird_record_index-1], bird_record_index, all_tokens[SAMPLE_INDEX][bird_record_index], "2021", bird_data[bird_record_index]))
            
    return db_records
```

```python
def store_records_in_db(db_records):
    conn = sqlite3.connect("./data/bird_lupu.db")
    sql = ''' INSERT INTO Birds (Location, BirdName, Month,Week,Sample, Year, Value) VALUES(?,?,?,?,?,?,?)'''  
    cur = conn.cursor()
    cur.executemany(sql,db_records)
    conn.commit()
```

```python
import os
all_tokens = []
for subdir, dirs, files in os.walk('./data/'):
    for f in files:
        if f[-3:] == 'txt': #only process txt files
            print(os.path.join(subdir, f))
            db_records = get_records_from_file(os.path.join(subdir, f))
            store_records_in_db(db_records)
```
